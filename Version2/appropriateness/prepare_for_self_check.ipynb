```python
import pandas as pd
import random
```

```python
df = pd.read_csv("Inappropriate_095_top_vs_one_with_multi.csv")
```

```python
df.head()
```

```python
len(df.columns[3:-3])
```

```python
topics = df.columns[3:-3]
```

```python
collected_df_list = []
for t in topics:
    for label in ['yes','no']:
        df_curr = df[(df[t]==1)&(df['OUTPUT:result']==label)]
        df_curr = df_curr.sample(5)
        df_curr = df_curr[['INPUT:text','OUTPUT:result']]
        topic_column = [t] * len(df_curr)
        df_curr['topic'] = topic_column
        collected_df_list.append(df_curr)
        
collected_df_list = pd.concat(collected_df_list)
collected_df_list.drop_duplicates(subset = ['INPUT:text'], inplace = True)
```

```python
collected_df_list.head()
```

```python
collected_df_list.to_csv("inapp_for_labeling.csv", index = None)
```
